```python
class stu:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.skills = []

    def add_skill(self, skill):
        return self.skills.append(skill)

    def info(self):
        return f"Name = {self.name}, Age = {self.age}, Skills = {self.skills}"
```

# Inheritance

```python
class person:
    insti = "Emerson"
    
    def __init__(self,age,ht):
        self.age = age
        self.ht = ht
        
    def show(self):
        print(f"age -> {self.age}, Hight -> {self.ht},institude ->{self.insti}")
```

```python
class Ravi(person):
    
    def __init__(self,age,ht,sal):
        super().__init__(age,ht)
        self.sal = sal
        
    def sal(self):
        print(f"salary is {self.sal}")
    
    
```

```python
a= Ravi(23,5.2 ,30000)
```

```python
a.sal
```

```python
a.show()
```

```python
# simple menu driven with oop
# add item ,view,exit(grocery list) 
```

```python
class menuapp:
    def __init__(self):
        self.itemlist =[]
        
    def show_menu(self):
        
        print("Menu \n 1.add_item \n2.view item\3.exit")
    
    def add_item(self):
        item = input("enter a item name")
        self.itemlist.append(item)
        
    def view_item(self):
        return self.itemlist
    
    def run(self):
        
        while True:
            self.show_menu()
            ch = int(input("enter the choice"))
            if ch==1:
                self.add_item()
            elif ch==2:
                self.view_item()
            elif ch==3:
                break
            else :
                return "invalid choice"
                
```

```python
sukanya= menuapp()
```

```python
sukanya.add_item()
```

```python
sukanya.view_item()
```

```python
samruddhi= menuapp()
```

```python
samruddhi.run()
```

```python
samruddhi.view_item()
```
