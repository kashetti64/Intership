```python
class person:
    def __init__(self,name,age,ht):
        self.name = name
        self.__age = age # to private __ is used
        self.height = ht # to protected _ single underscore is used
    
    def show_info(self):
        return f"name ->{self.name},age->{self.__age},height->{self.height}"
```

```python
p1 = person("sukanya",18,5.2)
p2 = person("samruddhi",17,5)
```

```python
p1.name
```

```python
p2.age
```

```python
p1.show_info()
```

```python
p2.show_info()
```
