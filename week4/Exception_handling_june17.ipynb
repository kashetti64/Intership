```python
try:
    f= open('demo.txt')
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
```

```python
try:
    f= open('demo.txt')
    if f.name =='demo.txt':
        raise Exception
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally :
    print("run complete")
```

```python
try :
    num1 = int(input("enter number 1->"))
    num2 = int(input("enter a number num2->"))
    result= num1 / num2
    print("result = ",result)
except ZeroDivisionError :
    print("cannot be divided")
except ValueError :
    print("wrong input")
```
