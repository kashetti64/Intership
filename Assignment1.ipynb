```python
year = int(input("Enter a year: "))
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("It is leap year(century leap)")
        else:
            print("century but not leap")
    
    else:
        print("It is leap year")

else:
    print("It is not leap year")
```
