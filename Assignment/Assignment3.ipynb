# Section A — Classes and Objects

```python
"""
1. Create a class BankAccount with the following:
a) Attributes: account_holder, account_number, and balance.
b) Methods: deposit(amount), withdraw(amount), and get_balance().
c) withdraw() should print a message and do nothing if the amount exceeds the current balance.
d) A __str__ method that returns a readable summary of the account.
Create two BankAccount objects and perform a few deposits and withdrawals on each.
""";
```

```python
class BankAccount :
    
    def __init__(self,account_holder,account_number,balance):
        self.account_holder = account_holder
        self.account_number =account_number
        self.balance = balance 
    def deposit(self,amount):
        amount = int(input("enter a amount to deposite"))
        self.balance+= amount
        print(f"Deposited ₹{amount}")
    def withdraw(self,amount):
        amount = int(input("enter a amount to withdraw"))
        if amount > self.balance:
            print("Insufficient balance!")
        else:
            self.balance -= amount
            print(f"Withdrawn ₹{amount}")
    def get_balance(self):
        print("Current Balance:", self.balance)
    def __str__(self):
        return (f"Account Holder: {self.account_holder}\n"
                f"Account Number: {self.account_number}\n"
                f"Balance: ₹{self.balance}")
acc1 = BankAccount("Aishwarya", 101, 10000)
acc2 = BankAccount("Sukanya", 102, 5000)

acc1.deposit(2000)
acc1.withdraw(1500)

acc2.deposit(1000)
acc2.withdraw(7000) 


print("\nAccount 1 Details:")
print(acc1)

print("\nAccount 2 Details:")
print(acc2)
```

```python
"""
2. Create a class Student with attributes name, roll_no, and marks (a list of 5 subject marks).
Add methods:
a) calculate_percentage() — returns the average of all marks.
b) get_grade() — returns 'A' for >= 80, 'B' for >= 60, 'C' for >= 40, and 'Fail' otherwise.
c) display() — prints all details including percentage and grade.
Create a list of 3 Student objects and call display() on each.
""";
```

```python
class student:
    def __init__(self,name,roll_no,marks):
        self.name =name
        self.roll_no = roll_no
        self.marks  = marks
        
    def calculate_per(self):
        return sum(self.marks)/len(self.marks)
    
    def get_grade(self):
        average = self.calculate_per()
                   
        if average >= 80 :
            return "A"
        elif average >= 60:
            return "B"
        elif average >= 40:
            return "C"
        else :
            return "Fail"
            
                   
    def display(self):
        print("Name = ",self.name)
        print("Rollno = ",self.roll_no)
        print("calculated percentage = ",self.calculate_per())
        print("Marks = ",self.marks)
        print("Grade = ",self.get_grade())
                   
stu = []

for i in range(3):
    print(f"\nEnter details of student {i+1}")

    name = input("Enter name: ")
    roll_no = input("Enter roll number: ")

    marks = []

    for j in range(5):
        mark = int(input(f"Enter mark {j+1}: "))
        marks.append(mark)

    s = student(name, roll_no, marks)
    stu.append(s)

for s in stu:
    s.display()
```

# Section B — Encapsulation

```python
"""
3. Create a class Employee with private attributes __name, __emp_id, and __salary. Add:
a) Getter and setter methods for all three attributes.
b) The salary setter should raise a ValueError if the value is below 10000.
c) A method give_raise(percent) that increases salary by the given percentage.
Demonstrate that the private attributes cannot be accessed directly, and that all access goes
through the methods.
""";
```

```python
class Employee():
    
    def __init__(self,name,emp_id,sal):
        self.__name = name
        self.__emp_id = emp_id
        self.__sal = sal
      #getter  
    def get_name(self):
        return self.__name
    
    def get_emp_id(self):
        return self.__emp_id
    
    def get_sal(self):
        return self.__sal
    # setter
    def set_name(self,name):
        self.__name = name
        
    def set_emp_id(self,emp_id):
        self.__emp_id = emp_id 
        
    def set_sal(self,sal):
        if sal < 10000:
            raise ValueError("Salary cannot be below 10000")
        self.__sal = sal
        
        
    def give_raise(self,percent):
        self.__sal += self.__sal *percent /100
        
e1 = Employee("sukanya",101,1800)
print("Name:", e1.get_name())
print("ID:", e1.get_emp_id())
print("Salary:", e1.get_sal())
e1.give_raise(10)
print("Salary after raise:", e1.get_sal())
try:
    e1.set_sal(5000)
except ValueError as e:
    print(e)
```

```python
"""
4. Build a class SecureVault that stores a private __pin and a private __data dictionary.
It should have:
a) A method add_entry(pin, key, value) that adds to __data only if the pin matches.
b) A method get_entry(pin, key) that returns the value only if the pin matches.
c) Any wrong pin attempt should print 'Access Denied' and return nothing.

""";
```

```python
class SecureVault:

    def __init__(self, pin):
        self.__pin = pin
        self.__data = {}

    def add_entry(self, pin, key, value):
        if pin == self.__pin:
            self.__data[key] = value
        else:
            print("Access Denied")

    def get_entry(self, pin, key):
        if pin == self.__pin:
            return self.__data.get(key)
        else:
            print("Access Denied")


# Create object
vault = SecureVault("1234")

# Add entries
vault.add_entry("1234", "name", "Sukanya")
vault.add_entry("1234", "city", "Pune")

# Get entries
print(vault.get_entry("1234", "name"))   # Sukanya
print(vault.get_entry("1234", "city"))   # Pune

# Wrong PIN
print(vault.get_entry("1111", "name"))   # Access Denied
```

# Section C — Inheritance and Method Overriding

```python
"""
5. Create a base class Shape with a method area() that returns 0 and a method describe() that
prints 'I am a shape'. Create three subclasses:
a) Circle(radius) — overrides area() using pi * r^2.
b) Rectangle(length, width) — overrides area().
c) Triangle(base, height) — overrides area().
Each subclass should also override describe() to print something specific about itself. Create one
object of each and call both methods.
""";
```

```python
import math 
#base class
class shape():
    def area(self):
        return 0
    
    def describe():
        print("I am shape")
#circle class
class circle(shape):
    def __init__(self,radius):
        self.radius = radius
        
    def area(self):
        return math.pi * self.radius *self.radius
    
    def describe(self):
        print("I am circle")
        
class rectangle(shape):
    def __init__(self,l,b):
        self.l = l
        self.b = b
    def area(self):
        return self.l * self.b
    def describe(self):
        print("I am rectangle")
        
class triangle(shape):
    def __init__(self,base,height):
        self.base = base
        self.height = height
    
    def area(self):
        return 0.5 * self.base * self.height
    
    def describe(self):
        print("I am Triangle")
        
c = circle(5)
r = rectangle(4,6)
t = triangle(5,8)
    
c.describe()
print("Area:", c.area())

r.describe()
print("Area:", r.area())

t.describe()
print("Area:", t.area())
```

```python
"""
6. Create a class Vehicle with attributes make, model, and year and a method start_engine() that
prints 'Engine started'. Create two subclasses:
a) Car(make, model, year, num_doors) — overrides start_engine() to include the car's name in
the message.
b) ElectricCar(make, model, year, battery_capacity) — overrides start_engine() to print 'Silent
start — battery at X kWh'.
Use super() in both subclasses to call the parent __init__. Create objects and test all methods.
""";
```

```python
class vehicle():
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year
    def start_engine(self):
        print("Engine started")
        
class car(vehicle):
    def __init__(self,make,model,year,num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors
    def start_engine(self):
        print(f"Engine started for {self.make} {self.model}")

class electric_car(vehicle):
    def __init__(self,make,model,year,battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity

    def start_engine(self):
        print(f"Silent start — battery at {self.battery_capacity} kWh")
    
#m= bike(20)
#c= car(2,34c,2026,4)
#= electric_car(22,735cb,100watt)

c= car("Toyota", "Camry", 2024, 4)
e = electric_car("Tesla", "Model 3", 2025, 75)

# Testing
c.start_engine()
e.start_engine()
```

# Section F — Menu-Driven App with OOP

```python
"""
10. Build a menu-driven Student Record Manager using OOP and file handling. The app should
have a class StudentManager with the following menu options:
1. Add Student — takes name, roll number, and marks for 3 subjects.
2. View All — displays all stored students with their average marks.
3. Search Student — finds a student by roll number.
4. Delete Student — removes a student by roll number.
5. Save to File — writes all records to students_data.txt.
6. Load from File — reads records back from students_data.txt on startup.
7. Exit
Requirements: use a class with a list as internal storage, handle invalid menu choices gracefully,
and use a while loop to keep the menu running until the user chooses to exit
""";
```

```python
class StudentManager:

    def __init__(self):
        self.students = []

    def add_stu(self):
        name = input("Enter Name: ")
        roll = input("Enter Roll No: ")

        marks = []

        for i in range(3):
            mark = int(input(f"Enter mark {i+1}: "))
            marks.append(mark)

        student = {
            "name": name,
            "roll": roll,
            "marks": marks
        }

        self.students.append(student)

    def view_all(self):
        for student in self.students:
            avg = sum(student["marks"]) / len(student["marks"])

            print(student["name"])
            print(student["roll"])
            print("Average =", avg)

    def search_stu(self):
        search = input("Enter Roll No: ")

        for student in self.students:
            if student["roll"] == search:
                print(student)
                return

        print("Student Not Found")

    def delete(self):
        roll = input("Enter Roll No: ")

        for student in self.students:
            if student["roll"] == roll:
                self.students.remove(student)
                print("Deleted")
                return

        print("Student Not Found")
```

```python
obj = StudentManager()

while True:

    print("1.Add")
    print("2.View")
    print("3.Search")
    print("4.Delete")
    print("5.Exit")

    choice = int(input("Enter Choice: "))

    if choice == 1:
        obj.add_stu()

    elif choice == 2:
        obj.view_all()

    elif choice == 3:
        obj.search_stu()

    elif choice == 4:
        obj.delete()

    elif choice == 5:
        break

    else:
        print("Invalid Choice")
```
