```python
"""
1. Write a function analyze_sentence(sentence) that takes a sentence and returns a dictionary with
word_count, char_count (excluding spaces), unique_words (case-insensitive), and
most_frequent_word. Test it with:
"The quick brown fox jumps over the lazy dog the fox"
""";
```

```python
def ana(a):
    word =a.lower().split()
    
    word_count = len(word)
    char_count = len(a.replace(" ", ""))
    unique = len(set(word))
    freq={}
    
    
    for i in word:
        freq[i] = freq.get(i, 0) + 1
        
    most_frequent_word = max(freq, key=freq.get)
    return {
            "word count" :word_count,
            " char_count":char_count,
            " unique":unique,
            "most_frequent_word": most_frequent_word
            }

a="The quick brown fox jumps over the lazy dog the fox"
result =ana(a)
print(result)
```

```python
"""
2. Given the string below, do the following and print each result:
text = "A man a plan a canal Panama"
a) Reverse the string without using slicing.
b) Check if it is a palindrome, ignoring case and spaces.
c) Replace every vowel with *.
d) Count how many times each vowel appears.

""";
```

```python
text = "A man a plan a canal Panama"

# a) Reverse the string
rev = ""
for i in text:
    rev = i + rev
print("Reverse =", rev)
# b) Palindrome check (ignore spaces and case)
s = text.replace(" ", "").lower()
rev2 = ""
for i in s:
    rev2 = i + rev2
if s == rev2:
    print("Palindrome")
else:
    print("Not Palindrome")
# c) Replace vowels with *
new_text = ""

for i in text:
    if i.lower() in "aeiou":
        new_text += "*"
    else:
        new_text += i

print("After replacing vowels =", new_text)
# d) Count vowels
a = e = i = o = u = 0
for ch in text.lower():
    if ch == "a":
        a += 1
    elif ch == "e":
        e += 1
    elif ch == "i":
        i += 1
    elif ch == "o":
        o += 1
    elif ch == "u":
        u += 1

print("a =", a)
print("e =", e)
print("i =", i)
print("o =", o)
print("u =", u)
```

```python
"""
7. Write a function inventory_manager(records) that takes a list of tuples (product, category,
quantity, price) and returns a dictionary grouped by category. Each category should have
total_items, total_value, and a list of product names. Use:
records = [ ("Laptop", "Electronics", 10, 55000), ("Mouse", "Electronics", 50,
499), ("Desk", "Furniture", 8, 12000), ("Chair", "Furniture", 15, 4500),
("Pen", "Stationery", 200, 20), ]


""";
```

```python
def inventory(records):
    inventory={}
    for pro ,cate,quantity,price in records:
        if cate not in inventory :
            inventory[cate]={
                "total item":0,
                "total value":0,
                "product" :[]
            }
        inventory[cate]["total item"] += quantity
        inventory[cate]["total value"] += quantity *price
        inventory[cate]["product"].append(pro)
    return inventory

records =[
    ("Laptop", "Electronics", 10, 55000), ("Mouse", "Electronics", 50,499),
    ("Desk", "Furniture", 8, 12000), ("Chair", "Furniture", 15, 4500),
    ("Pen", "Stationery", 200, 20), 
]
result = inventory(records)

for cate, details in result.items():
    print(cate)
    print(details)
    print()
```

```python
# a) power_all(*args, exp=2)
def power_all(*args, exp=2):
    result = []

    for i in args:
        result.append(i ** exp)

    return result


print(power_all(2, 3, 4))
print(power_all(2, 3, 4, exp=3))
# b) profile(**kwargs)

def profile(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)
profile(name="Aishwarya", age=19, city="Pune")
# c) Sort by cgpa in descending order
students = [
    {"name": "Arjun", "cgpa": 8.4},
    {"name": "Sanya", "cgpa": 9.1},
    {"name": "Dev", "cgpa": 7.8}
]

sorted_students = sorted(
    students,
    key=lambda x: x["cgpa"],
    reverse=True
)

print(sorted_students)
```

```python
"""
9. Create a module stats_utils.py with three functions:
a) mean(data) — arithmetic mean.
b) variance(data) — population variance without using the statistics module.
c) normalize(data) — min-max normalization returning values between 0 and 1.
In a separate main.py, import and use these functions on the dataset below. Also compute the
standard deviation using the math module.
dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]
""";
```

```python
%%writefile stats_utils.py
def mean(data):
    return sum(data) / len(data)

def variance(data):
    avg = mean(data)

    total = 0
    for x in data:
        total += (x - avg) ** 2

    return total / len(data)      # Population variance

def normalize(data):
    min_val = min(data)
    max_val = max(data)

    normalized = []
    for x in data:
        norm = (x - min_val) / (max_val - min_val)
        normalized.append(norm)

    return normalized
```

```python
import math
from stats_utils import mean, variance, normalize

dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]

avg = mean(dataset)
var = variance(dataset)
std_dev = math.sqrt(var)
norm_data = normalize(dataset)

print("Dataset:", dataset)
print("Mean:", avg)
print("Variance:", var)
print("Standard Deviation:", std_dev)
print("Normalized Data:", norm_data)
```

```python
"""
10. Complete the following:
a) Write a closure make_multiplier(n) that returns a function multiplying its input by n. Create triple
and quintuple using it.
b) Write a decorator @timer that prints the execution time of any function it wraps. Apply it to
slow_sum(n) which computes the sum of squares from 1 to n.
c) Write a decorator @validate_positive that raises a ValueError if any argument is zero or negative.
Apply it to safe_sqrt(x).
""";
```

```python
#a
def make_multiplier(n):
    def multiplier(x):
        return x * n
    return multiplier

# Creating functions
triple = make_multiplier(3)
quintuple = make_multiplier(5)

print("Triple of 10 =", triple(10))
print("Quintuple of 10 =", quintuple(10))
```

```python
# B
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()

        result = func(*args, **kwargs)

        end = time.time()
        print("Execution Time:", end - start, "seconds")
        return result
    return wrapper

@timer
def slow_sum(n):
    total = 0
    for i in range(1, n + 1):
        total += i ** 2
    return total

print("Sum =", slow_sum(100000))
```

```python
#c
import math

def validate_positive(func):
    def wrapper(*args, **kwargs):
        for arg in args:
            if arg <= 0:
                raise ValueError("Arguments must be positive")

        return func(*args, **kwargs)
    return wrapper

@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)

print(safe_sqrt(25))

# Test with invalid value
print(safe_sqrt(9))
```
