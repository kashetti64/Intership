# Section A — Exception Handling

```python
"""
1. Write a function safe_divide(a, b) that divides two numbers and handles ZeroDivisionError and
TypeError separately, printing an appropriate message for each. Use a finally block to print
'Operation attempted.' regardless of the outcome. Test it with valid numbers, division by zero, and a
string input.

""";
```

```python
def safe_divide(a,b):
    try:
        c= a/b
        print(c)
        
    except ZeroDivisionError :
        print("Error : zdivision not possiable")
        
    except TypeError :
        print("wrong input")
        
    finally :
        print("Operation attempted.")
```

```python
safe_divide(2,"10")# String input
safe_divide(13,0)  # Division by zero
safe_divide(10,2) #valid input
```

```python
"""
2. Write a function validate_age(age) that raises a custom exception InvalidAgeError (a class you
define, inheriting from Exception) if age is negative or greater than 120. Catch this exception in a
calling block and print a user-friendly message. Test it with at least 3 different values.
""";
```

```python
class Age(Exception):
    pass 
def validate_age(age):
    if age < 0 or age >120 :
        raise Age("Age must be between 0 and 120.")
    print(f"Age = {age} is valid ")
    
test_ages = [25, 130, -5]

for age in test_ages:
    try :
        validate_age(age)
    except Age as e :
        print(f"Invalid age {age}: {e}")
```

```python
"""
3. Write a program that repeatedly asks the user (use a loop with at least 3 simulated inputs in a list,
e.g. user_inputs = ['10', 'abc', '0', '5']) to enter a number and divide 100 by it. Handle the following in
a single try block using multiple except clauses:
a) ValueError if the input is not a number.
b) ZeroDivisionError if the input is zero.
c) A generic Exception as a catch-all for anything else.
Print the result if successful, otherwise print which error occurred.
""";
```

```python
user_inputs = ['10', 'abc', '0', '5']
for value in user_inputs :
    try :
        num = int(value)
        result = 100/num
        print(f"100 / {num} = {result}")
    except ValueError:
        print(f"The {value} you entered is not int")
    except ZeroDivisionError :
        print("division by zero is not possiable")
    except Exception as e :
        print(f"unexpected error:{e}")
```

```python
np.random.seed(42)
arr = np.random.randint(1,101,15)
print(arr)
# A
print("Mean = ",arr.mean())
print("Median = ",np.median(arr))
print("std = ",arr.std())
print("Min = ",arr.min())
print("Max = ",arr.max())

#B
des = np.sort(arr[::-1])
print("desending= ",des)

#c
top3_indices = np.argsort(arr)[-3:][::-1]

print("Top 3 indices:", top3_indices)
print("Top 3 values:", arr[top3_indices])

#D # revise
normalized = (arr - arr.min()) / (arr.max() - arr.min())

print("Normalized array:", normalized)
```

```python
"""
7. Create a Pandas DataFrame from the dictionary below and perform the following:
data = { "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"], "Age": [23, 21,
25, 22, 24], "Department": ["AI", "Data Science", "AI", "Web Dev", "Data
Science"], "Score": [88, 92, 79, 95, 67], }
a) Display the first 3 rows and the data types of each column.
b) Add a new column Grade based on Score (>=90 'A', >=75 'B', else 'C').
c) Sort the DataFrame by Score in descending order.
d) Filter rows where Department is 'AI'.
""";
```

```python
import pandas as pd
data = { "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"], 
        "Age": [23, 21,25, 22, 24],
        "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
        "Score": [88, 92, 79, 95, 67], }
df = pd.DataFrame(data)
# A
print(df.head(3))
print(df.dtypes)
#B
def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 75:
        return "B"
    else:
        return "C"

df["Grade"] = df["Score"].apply(get_grade)

#C
sort_df = df.sort_values(by="Score",ascending = False)
print(sort_df)
#D
ai_df = df[df["Department"] == "AI"]
print(ai_df)
```

```python
data = { "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"], 
        "Age": [23, 21,25, 22, 24],
        "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
        "Score": [88, 92, 79, 95, 67], }
df= pd.DataFrame(data)
print(df)
#A
avg = df.groupby("Department")["Score"].mean()
print("\naverage=",avg)
#b
high_score= df.loc[df["Score"].idxmax()]

print("\nRow with Highest Score:")
print(high_score)
#c
rename = df.rename(columns= {"Score": "Marks"})
print("\nData feames afte renaming\n",rename)

#D
df_dropped = df.drop(columns = ["Age"])
print("\nAfter deleting Age\n",df_dropped)
```

```python
"""
9. Create a DataFrame with some missing values as shown below and handle them:
import numpy as np data = { "Product": ["Pen", "Notebook", "Eraser",
"Sharpener", "Ruler"], "Price": [10, np.nan, 5, 8, np.nan], "Stock": [100, 50,
np.nan, 30, 20], }
a) Check the number of missing values in each column using isnull().sum().
b) Fill missing Price values with the column mean.
c) Drop rows where Stock is still missing.
d) Save the cleaned DataFrame to a CSV file named cleaned_products.csv without the index
""";
```

```python
import pandas as pd
import numpy as np

data = {
    "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
    "Price": [10, np.nan, 5, 8, np.nan],
    "Stock": [100, 50, np.nan, 30, 20]
}

df = pd.DataFrame(data)

# a) Count missing values
print(df.isnull().sum())

# b) Fill missing Price values with mean
df["Price"] = df["Price"].fillna(df["Price"].mean())

# c) Drop rows where Stock is missing
df = df.dropna(subset=["Stock"])

# d) Save cleaned DataFrame to CSV
df.to_csv("cleaned_products.csv", index=False)

print(df)
```
