```python
# create a function to find factorial of number
```

```python
def fact(n):
    fact =1
    for i in range(1, n + 1) :
        fact = fact * i
    print(fact)
    
```

```python
# pass any number of integer in a function and find their sum
```

```python
def addall(*args):
    total = 0
    
    for i in args:
        total = total + i
        
    return total
```

```python
addall(11,34,22,1)
```
