```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import seaborn as sns

# Load dataset
df = sns.load_dataset('taxis')

# Remove pickup and dropoff columns
df = df.drop(columns=['pickup', 'dropoff'])

# Function to encode all categorical columns
def encode(df):
    le = LabelEncoder()

    for column in df.columns:
        if df[column].dtype == 'object' or str(df[column].dtype) == 'category':
            df[column] = le.fit_transform(df[column])

    return df

# Encode the dataset
newdf = encode(df)

print(newdf.head())
```
