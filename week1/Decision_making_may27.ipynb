```python
year = int(input("Enter a year: "))
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("It is leap year(century leap)")
        else:
            print("It is not leap year")
    
    else:
        print("It is leap year")

else:
    print("It is not leap year")
```

```python
stri = str(input("enter a string "))
print("Menu => /n 1.Uppercase /n2.lowercase/n3.replace part with given part")
ch=int(input("enter option from 1 to 3rd"))
if ch==1 :
    print("string",stri.upper())
elif ch ==2 :
    print("Lowercase",stri.lower())
elif ch==3 :
    r1 = input("which part u want to replace")
    r2 = input("what you want to insert:=")
    new=stri.replace(r1,r2)
    print(f"result{new}")
```
