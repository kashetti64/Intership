```python
"""
Vowel Counter:
Write a Python function that takes a string as input. Using a for loop, 
iterate through each character. If the character is a vowel 
(a, e, i, o, u, case-insensitive),
increment a counter. Store the count of each vowel in a 
dictionary. Finally, print
the total number of vowels found and the dictionary of vowel counts
""";
```

```python
def vowel(a):
    count=0
    vowel ={'a':0,'e':0,'i':0,'o':0,'u':0}
    for i in a:
        i = i.lower()
        if i in vowel:
            count =+1
            vowel[i] =+1
    print(count)
    print(vowel)
```

```python
"""
Palindrome Checker with List:
Write a Python function that accepts a string. Inside the function, 
use a for loop to build a new list containing the characters of the string 
in reverse order. Then, use an if-else statement to check if the original
string is a palindrome (reads the same forwards and backward) by comparing
it to the string formed from the reversed list. Print whether the string 
is a palindrome or not.
""";
```

```python
def pali(a):
    rev_list = []

    for i in range(len(a)-1, -1, -1):
        rev_list.append(a[i])

    pal1 = "".join(rev_list)

    if a == pal1:
        print("Palindrome")
    else:
        print("Not a palindrome")

text = input("Enter a string: ")
pali(text)
```

```python
"""
4. Word Length Classication:
Write a Python function that takes a sentence (string) as input. Split the sentence
into words. Using a for loop, iterate through each word. Use if-elif-else
statements to classify words as "short" (length < 5), "medium" (5 <= length <= 8),
or "long" (length > 8). Store these classied words in separate lists (e.g.,
short_words, medium_words, long_words). Print all three lists.

""";
```

```python
def leng(s) :
    short_words=[]
    medium_words=[]
    long_words=[]
    word = s.split()
    
    for i in word:
        if len(i) <5 :
            short_words.append(i)
        elif len(i) <= 8:
            medium_words.append(i)
        else :
            long_words.append(i)
            
    print("Short words :", short_words)
    print("Medium words:", medium_words)
    print("Long words  :", long_words)
```

```python
""" 6:=
Unique Word Extractor:
Write a function that takes a paragraph (string) as input. Convert the paragraph
to lowercase and remove punctuation. Use a for loop to split the paragraph into
words and add each unique word to a set data structure. Finally, convert the set
back to a sorted list and print it.
""";
```

```python
def unique_words(paragraph):
    paragraph = paragraph.lower()

    for ch in ".,!?":
        paragraph = paragraph.replace(ch, "")

    s = set()

    for word in paragraph.split():
        s.add(word)

    print(sorted(list(s)))

para = input("Enter paragraph: ")
unique_words(para)
```

```python
"""
8. String Anagram Checker:
Write a function that takes two strings as input. Convert both strings into lists of
characters. Sort both lists. Use an if-else statement to check if the sorted lists are
identical. Print whether the two strings are anagrams of each other (contain the
same characters with the same frequencies).
""";
```

```python
def anagram(s1, s2):
    l1 = list(s1.lower())
    l2 = list(s2.lower())

    l1.sort()
    l2.sort()

    if l1 == l2:
        print("Strings are anagrams")
    else:
        print("Strings are not anagrams")

anagram("listen", "silent")
```

```python
"""
9. Missing Character Finder (Alphabet Check):
Given a string, write a function that determines which lowercase English alphabet
leers are not present in the string. Use a set to store all lowercase leers. Iterate
through the input string, removing each found leer from the set. Print the
remaining leers in the set.

""";
```

```python
def missing_chars(text):
    alphabets = set("abcdefghijklmnopqrstuvwxyz")

    for ch in text.lower():
        if ch in alphabets:
            alphabets.remove(ch)

    print("Missing letters:", alphabets)

missing_chars("Write a Python function that accepts a string.")
```

```python
"""
10. Custom String Reversal (Word by Word):
Write a function that takes a sentence. Use a for loop to iterate through the words
of the sentence. Store each word in a list. Then, using another loop (or list slicing),
reverse the order of words in the list. Finally, join the reversed words back into a
single string and print the result. (Example: "Hello World" -> "World Hello")
""";
```

```python
def reverse_words(sentence):
    words = []

    # First loop: store words in a list
    for word in sentence.split():
        words.append(word)

    rev = []

    # Second loop: reverse the list
    for i in range(len(words)-1, -1, -1):
        rev.append(words[i])

    result = " ".join(rev)
    print(result)

reverse_words("Hello World")
```
