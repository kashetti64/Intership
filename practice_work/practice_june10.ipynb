```python
"""
 8: Merge and Sort Two Lists (Manual Sort)
Given two lists:
● list_a = [3, 1, 4]
● list_b = [2, 5, 0]
Combine these two lists into a single list and then sort the combined list in
ascending order without using the list.sort() method or sorted() function. You'll
need to implement a basic sorting algorithm (e.g., bubble sort logic) using loops.
""";
```

```python
# revise
list_a = [3, 1, 4]
list_b = [2, 5, 0]

combined = list_a + list_b

for i in range(len(combined)):
    for j in range(len(combined)-1):
        if combined[j] > combined[j+1]:
            temp = combined[j]
            combined[j] = combined[j+1]
            combined[j+1] = temp

print(combined)
```

```python
"""
Question 9: Flaen a Nested List
You are given a list that contains integers and other lists (only one level of nesting). For
example: nested_list = [1, [2, 3], 4, [5, 6, 7], 8].
Write code to create a new at list containing all the integers from nested_list. The result
should be [1, 2, 3, 4, 5, 6, 7, 8].
""";
```

```python
nested_list = [1, [2, 3], 4, [5, 6, 7], 8]
new_list = [ ]

for i in nested_list:
    if type(i) == list:
        for j in i:
            new_list.append(j)
    else:
        new_list.append(i)

print(new_list)
```

```python
"""
Question 10: Find Common Elements
Given two lists, list_p = [1, 2, 3, 4, 5] and list_q = [4, 5, 6, 7, 8], write code to nd and print all
elements that are common to both lists. The order of common elements does not maer. You
are not allowed to use set operations.
""";
```

```python
list_p = [1, 2, 3, 4, 5]
list_q = [4, 5, 6, 7, 8]

result = []

for i in list_p:
    result.append(i)

for i in list_q:
    if i not in result:
        result.append(i)

print(result)
```
